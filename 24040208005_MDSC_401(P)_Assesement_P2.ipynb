```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, accuracy_score
from sklearn.naive_bayes import  MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
 
```

```python
df = pd.read_csv(r"E:\MSC 4\MLOps\Churn_Modelling.csv")
```

```python
df.head()
```

```python
labelencoder_geography = LabelEncoder()
labelencoder_gender = LabelEncoder()

df['Geography'] = labelencoder_geography.fit_transform(df['Geography'])
df['Gender'] = labelencoder_gender.fit_transform(df['Gender'])
```

```python
df = df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
```

```python
X = df.drop(columns=["Exited"])
Y = df['Exited']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 42)
```

```python
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

Model 1 : Logistic Regression as classifier

```python
model1 = LogisticRegression()

model1.fit(X_train, Y_train)

Y_pred = model1.predict(X_test)
acc = accuracy_score(Y_test, Y_pred)

print(f"LogisticRegression has accuracy: {acc}")
```
